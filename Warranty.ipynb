```python
import pandas as pd
from neo4j.v1 import GraphDatabase, basic_auth
from IPython.display import display

driver = GraphDatabase.driver("bolt://localhost:7687", auth=basic_auth("neo4j", "SwAbMp14"))
session = driver.session()
# result = session.run("""
# MATCH (a:Word )-[r:HasNextWord{ClaimNumber: 'X004027', SentenceNumber: 3}]->(b:Word)
# RETURN a.name, type(r), b.name LIMIT 10;
# """)

```

```python
result = session.run("""
MATCH (a)-[r:HasNextWord {
ClaimNumber:'A329354'}]->(b) 
RETURN a.name, ID(a), type(r), r.SequenceNumber, r.SentenceNumber, r.ClaimNumber,  b.name, ID(b)
ORDER BY toInteger(r.SentenceNumber) ASC, 
toInteger(r.SequenceNumber) ASC ;
""")

main_df = pd.DataFrame(columns=['claim_no', 'a_name', 'a_id', 'relation', 'sequence_no', 'sentence_no', 'b_name', 'b_id'])

counter = 0
for record in result:
#     print(record)
    main_df.loc[counter] = [record["r.ClaimNumber"], 
                            record["a.name"], record["ID(a)"], 
                            record["type(r)"], 
                            record["r.SequenceNumber"], 
                            record["r.SentenceNumber"],
                            record["b.name"],
                            record["ID(b)"]]
    counter += 1
    
display(main_df)
```
