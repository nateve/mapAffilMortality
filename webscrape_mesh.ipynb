```python
def scrape_and_clean(url):
    print(url)
    res = requests.get(url)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')
    toplevel = soup.find('ul', {'class':'Level1'})
    list_all = toplevel.get_text().split('\n')
    print(list_all[0])
    list_all = list(filter(None, list_all))
    regex = re.compile(r'\[.*\]', re.IGNORECASE)
    clean_list = [regex.sub("", line).strip().lower() for line in list_all]
    return clean_list
    
```

```python
circulatory_list = scrape_and_clean(circulatory)
respiratory_list = scrape_and_clean(respiratory)
mental_list = scrape_and_clean(mental)
neoplasms_list = scrape_and_clean(neoplasms)
nervous_list = scrape_and_clean(nervous)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

fuzz.token_set_ratio("limbic encephalits", "encephlitis")

```

```python
mesh = pd.DataFrame(columns=['PMID','mesh'])
mesh['PMID'] = ['1001', '1002']
mesh['mesh'] = ['Atrial|carotid|encephalitis, varicella zoster|Encephalitis|restless legs syndrome', 'kartagener syndrome|alcohol withdrawal delirium|hypertrophy, left ventricular|parasystole']
```

```python

mesh['cleantext'] = mesh.mesh.str.lower().str.split('|')
#mesh['cleantext'] = mesh.cleantext.str.split('|')
```

```python
pd.set_option('max_colwidth',200)
mesh.head()
```

```python
def check_fuzzy(check_set, word):
    ratios = []
    for i in check_set:
        ratio = fuzz.token_set_ratio(i, word)
        ratios.append((i,ratio))
    maxval = max(ratios,key=lambda item:item[1])
    return maxval

#check_fuzzy(mental_list, 'hypertrophy')

def check_group(word):
    mesh_groups = set()
    sets = [circulatory_list,mental_list,nervous_list,respiratory_list,neoplasms_list]
    for group in sets:
        if word in set(group):
            #print(word, group[0])
            mesh_groups.add(group[0])
        elif check_fuzzy(set(group), word)[1] > 90:
            #print(check_fuzzy(set(group), word))
            mesh_groups.add(group[0])
        else:
            continue
            #print("no match",word)
    return(mesh_groups)

# check_group('neonatal')
```

```python
# list = [i for i in mesh.cleantext]
dicts = mesh.loc[:,['PMID', 'cleantext']].to_dict(orient='records')

for d in dicts:
    #print(d)
#     d['mesh_groups'] = set()
    for i in d['cleantext']:
        d['mesh_groups'] = check_group(i)
            

print(dicts)
```
