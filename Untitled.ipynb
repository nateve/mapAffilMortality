```python
%%time 

def plot_map(span, column):

    df = count_authors_delta.loc[count_authors_delta.time_span == span,[column, 'fips']].drop_duplicates()
    
    merged = pd.merge(df, county_df, on='fips', how='right')
    # merged_pad = merged.fillna(method='pad')
#     merged_na = merged.fillna(0)
    
    geo_data = [{'name': 'counties',
                 'url': county_topo,
                 'feature': 'us_counties.geo'}]
    
    vis = vincent.Map(data=merged, geo_data=geo_data, scale=2100, 
                      brew='Spectral',
                      projection='albersUsa', data_bind=column,
                      data_key='fips', map_key={'counties': 'properties.FIPS'})
    vis.marks[0].properties.enter.stroke_opacity = vincent.ValueRef(value=0.5)
    #Change our domain for an even inteager
    vis.scales['color'].type = 'threshold'
    highest = merged_na[column].max()
    lowest = merged_na[column].min()
    sevenths =  list(np.arange(lowest,highest, (highest-lowest)/7))
    sevenths.append(highest)
#     fifths = list(np.arange(min(merged_na[column]),sevenths[1],sevenths[1]/5))
#     fifths.append(max(merged_na[column]))
    print(sevenths)
#     print(fifths)
#     vis.scales['color'].domain = sevenths
    vis.scales['color'].domain = [-.1, -0.025, -0.0125, -0.00625, 0, 0.00625, 0.0125, 0.1]

    vis.legend(title='% authors published on '+column)
#     vis.to_json('vega.json')

    # vis.rebind(column='circulatory', brew='YlGnBu')
    # vis.scales['color'].domain = [0, 15]
    # vis.legends[0].title = 'proportion circulatory'
    
    #vis.display()
    return vis
```

```python
plot_map(1,'delta_resp')
```
