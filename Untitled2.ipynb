```python
import pandas as pd
```

```python
iter_csv = pd.read_csv('../data/mesh_groups_2016.tsv', sep='\t',
                         dtype={'PMID': object},
                         iterator=True, chunksize=10**6)

mdf = pd.concat([chunk for chunk in iter_csv])
```

```python
mdf.head()
```

```python
mdf.mesh_groups = mdf.mesh_groups.apply(lambda x: x if pd.isnull(x) else (str(x)[1:-1]))
```

```python
mdf.mesh = mdf.mesh.apply(lambda x: x if pd.isnull(x) else (str(x).replace("'", "")))
```
